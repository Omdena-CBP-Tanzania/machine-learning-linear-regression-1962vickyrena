### step 3:Data visualization

### Distribution of the house price

```python
# distribution of the  house price
plt.figure(figsize=(6,4))
sn.histplot(Boston_house["medv"], bins=20,kde=True)
plt.xlabel("Median  House Price (MEDV)")
plt.ylabel("Frequency")
plt.title("Distribution  of house price")
plt.show()
```

```python
# checking wheather the houses are near the river

plt.figure(figsize=(6,4))
sn.boxplot(x=Boston_house["chas"],y=Boston_house["medv"])
plt.xlabel("CHAR (0: not near  river,1: Near river)")
plt.ylabel("Median  House  Price (medv)")
plt.title("impact of CHAS on price")
plt.show()
```

```python
# predicting how  lower status population areas affect house price 
plt.figure(figsize=(6, 4))
sn.scatterplot(x=Boston_house['lstat'], y=Boston_house['medv'])
plt.xlabel("Lower Status Population (%)")
plt.ylabel("Median House Price (MEDV)")
plt.title("LSTAT vs MEDV")
plt.show()
```

```python
# showing  how crime rate affect  house price
plt.figure(figsize=(6, 4))
sn.scatterplot(x=Boston_house['crim'], y=Boston_house['medv'])
plt.xlabel("Crime Rate (CRIM)")
plt.ylabel("Median House Price (MEDV)")
plt.title("Crime rate  vs  House price")
plt.xscale('log')  # Log scale for better visualization
plt.show()
```

#### correlation Heatmap

```python
# Compute correlation matrix
correlation_matrix = Boston_house.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sn.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
